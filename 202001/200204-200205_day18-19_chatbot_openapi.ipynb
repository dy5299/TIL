## recursive function v.s. dictionary
피보나치 수열 45번째 구해봐  
dictionary : cash memory로 보면 돼

```python
#실패
def fiv(a1,a2,n):
    output = a1 * (n-3) + a2 * (n-2)
    return output

fiv(1,1,35)
```

```python
#pf - recursive function
def fibo(n):
    if n<3 :
        return 1
    else :
        return fibo(n-1) + fibo(n-2)
```

```python
print(fibo(10))
```

```python
print(fibo(40))
```

```python
print(fibo(45)) #3분 소요
```

```python
def fibo(a1,a2,n) :
    out = [a1,a2]
    for i in range(1,n+1):
        out.append(out[i]+out[i-1])

    return out[n-1]

print(fibo(1,1,40))
```

```python
#pf dictionary 이용 - 이미 계산된 건 또 계산하지 말자
pre = {1:1, 2:1}
def fibo2(n) :
    v = pre.get(n)
    if v != None :
        return v
    else :
        sum = fibo2(n-1) + fibo2(n-2)
        pre[n] = sum
        return sum
```

```python
print(fibo2(45))
print(fibo2(42))
print(pre)
```

### 함수로

```python
def getQuery(word) :
    url = 'https://search.naver.com/search.naver?where=kdic&query='
    url = url + urllib.parse.quote_plus(word)
    
    bs = BeautifulSoup(urllib.request.urlopen(url).read(), "html.parser")
    
    output = bs.select('p.txt_box')
    
    return [node.text for node in output]
```
