```python
# tensofow BMI 
import pandas as pd
import numpy as np
import tensorflow as tf

# , ,   CSV   
df = pd.read_csv("./dataset/bmi.csv")

#  
df["height"]=df["height"]/200
df["weight"]=df["weight"]/100

#   
bclass = {"thin" : [1,0,0], "normal":[0,1,0], "fat":[0,0,1]}
df["label_pat"]=df["label"].apply(lambda x: np.array(bclass[x]))

#    
test_df = df[15000:20000]
test_pat = test_df[["weight","height"]]
test_ans = list(test_df["label_pat"])

#   
X = tf.placeholder(tf.float32, [None,2]) #,    placeholder 
Y = tf.placeholder(tf.float32, [None,3]) #    placeholder 

W = tf.Variable(tf.zeros([2,3]))
b = tf.Variable(tf.zeros([3]))

y = tf.nn.softmax(tf.matmul(X,W)+b) #  
cross_entropy=tf.reduce_sum(Y*tf.log(y)) # -  
train = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy) # 

#,  
predict = tf.equal(tf.argmax(y,1), tf.argmax(Y,1))
accuracy = tf.reduce_mean(tf.cast(predict, tf.float32))

sess=tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(3500):
    i = (step*100) % 14000
    rows = df[i+1:i+1+100]
    x_pat = rows[["weight", "height"]]
    y_ans=list(rows["label_pat"])
    sess.run(train, feed_dict={X:x_pat, y:y_ans})
    if step%500 == 0:
        cre = sess.run(cross_entropy, feed_dict={X:x_pat, y:y_ans})
        acc = sess.run(accuracy, feed_dict={X:test_pat, y:test_ans})
        print("Epoch=", step, "=", cre, "()=", acc)
```

```python
###########tensofow bmi  ##################################
import pandas as pd
import numpy as np
import tensorflow as tf

df = pd.read_csv("./dataset/bmi.csv")

#  
df["height"] = df["height"] / 200
df["weight"] = df["weight"] / 100

#label   - thin[1, 0, 0]/normal[0, 1, 0]/fat [0, 0, 1]
bclass = {"thin": [1, 0, 0] , "normal":[0, 1, 0], "fat": [0, 0, 1]}
df["label_pat"] = df["label"].apply(lambda x: np.array(bclass[x]))

#   
test_df = df[15000:20000]
test_pat= test_df[["weight", "height"]]
test_ans = list(test_df["label_pat"])

X = tf.placeholder(tf.float32, [None, 2]) #,    placeholder  
Y = tf.placeholder(tf.float32, [None, 3])   #    placeholder  

W = tf.Variable(tf.zeros([2, 3])) 
b = tf.Variable(tf.zeros([3])) 

y = tf.nn.softmax(tf.matmul(X, W) + b)  #  
cross_entropy = -tf.reduce_sum(Y * tf.log(y))  # -  
train= tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)  # 
 
#,  
predict = tf.equal(tf.argmax(y, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(predict, tf.float32))

sess = tf.Session()
sess.run(tf.global_variables_initializer()) 
for step in range(3501):
    i = (step*100) % 14000
    rows = df[i+1:i+1+100]
    x_pat = rows[["weight", "height"]]
    y_ans =  list(rows["label_pat"])
    sess.run(train, feed_dict={X: x_pat  , Y: y_ans })
    if step%500  == 0 :
        cre = sess.run(cross_entropy, feed_dict={X: x_pat  , Y: y_ans })
        acc = sess.run(accuracy , feed_dict={X: test_pat  , Y: test_ans })
        print("Epoch=", step, "=", cre, "()=", acc)
```

```python
# keras + tensorflow  BMI 
import pandas as pd, numpy as np, tensorflow as tf
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.callbacks import EarlyStopping

# BMI    
df["height"] = df["height"] / 200
df["weight"] = df["weight"] / 100

#   
X = df[["weight", "height"]].as_matrix()

#label   - thin[1, 0, 0]/normal[0, 1, 0]/fat [0, 0, 1]
bclass = {"thin": [1, 0, 0] , "normal":[0, 1, 0], "fat": [0, 0, 1]}
Y = np.empty((20000,3))
for i, v in enumerate(df["label"]):
    Y[i] = bclass[v]

#       
X_train, Y_train = X[1:15001], Y[1:15001]
X_test, Y_test = X[15001:20001], Y[15001:20001]

#   
model = Sequential() #  
model.add(Dense(512, input_shape=(2,))) #Dense(  , ....)   
model.add(Activation('relu')) # 
model.add(Dropout(0.1)) # 
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.1))
model.add(Dense(3)) #       
model.add(Activation('softmax'))    # 

model.compile(loss="categorical_crossentropy", optimizer="rmsprop", metrics=['accuracy'])
#  : fit
hist=model.fit(X_train, Y_train, batch_size=100, nb_epoch=20, validation_split=0.1, callbacks=[EarlyStopping(monitor='val_loss', patience=2)], verbose=1)

#  : evaluate
score=model.evaluate(X_test, Y_test)
print("loss=", score[0])
print("accuracy=", score[1])


#weight decay( ) -           
# weight decay(  ) -           
# Dropout -                 
# softmax  -    0~1       1   
#                             
```

```python
import numpy as np

x = [2,4,6,8]
y = [81,93,91,97]

mx = np.mean(x)
my = np.mean(y)
print("x :", mx)
print("y :", my)

divisor = sum([(mx-i)**2 for i in x]) #a 

def top(x,mx,y,my) : #a 
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i]-my)
    return d

dividend = top(x,mx,y,my)
print(":", divisor)
print(":", dividend)

a = dividend / divisor
b = my - (mx*a)

print(":",a)
print("y:",b)
```

```python
import tensorflow as tf

# x, y  
data = [[2,81],[4,93],[6,91],[8,97]] #data set
x_data = [ i[0] for i in data ]
y_data = [ i[1] for i in data ]

#  a y  b   .
# ,   0 ~ 10 , y  0 ~ 100   .
a = tf.Variable(tf.random_uniform([1],0,10,dtype=tf.float64,seed=0))
b = tf.Variable(tf.random_uniform([1],0,100,dtype=tf.float64,seed=0))

# y    ax+b  .
y=a*x_data+b

#  RMSE 
rmse = tf.sqrt(tf.reduce_mean(tf.square(y-y_data)))

#  
learning_rate = 0.1

# RMSE     
gradient_descent = tf.train.GradientDescentOptimizer(learning_rate).minimize(rmse)

#   
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer()) #  , 
    for step in range(2001): #2000 
        sess.run(gradient_descent)
        if step % 100 == 0:
            print("Epoch: %.f, RMSE=%.4f,  a=%.4f,  b=%.4f" % (step, sess.run(rmse),sess.run(a), sess.run(b)))

# Epoch      
```

```python
import tensorflow as tf
import numpy as np

#  
w11 = np.array([-2,-2])
w12 = np.array([2,2])
w2 = np.array([1,1])

b1=3
b2=-1
b3=-1

# 
def MLP(x,w,b):
    y = np.sum(w*x)+b
    if y<=0:
        return 0
    else :
        return 1

# NAND 
def NAND(x1, x2):
    return MLP(np.array([x1,x2]), w11, b1)

# OR 
def OR(x1, x2):
    return MLP(np.array([x1,x2]), w12, b2)

# AND 
def AND(x1, x2):
    return MLP(np.array([x1,x2]), w2, b3)

# XOR 
def XOR(x1, x2):
    return AND(NAND(x1,x2), OR(x1,x2))

# x1, x2      
xs = [(0,0), (0,1), (1,0), (1,1)]
if __name__ == '__main__':
    for x in xs:
        y = XOR(x[0], x[1])
        print(" :"+ str(x)+"   :"+str(y))
```

```python
from keras.models import Sequential 
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy
import tensorflow as tf

seed = 0
numpy.random.seed(seed)        # seed  
tf.set_random_seed(seed)

df = pd.read_csv('./dataset/sonar.csv', header=None)           #  
dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)            #  


from sklearn.model_selection import StratifiedKFold

n_fold = 10  #10 
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)

accuracy = []  #  accuracy 

for train, test in skf.split(X, Y):  #  , , 
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error',  optimizer='adam',   metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs=100, batch_size=5)
    k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])
    accuracy.append(k_accuracy)

#  
print("\n %.f fold accuracy:" % n_fold, accuracy)
```
